```python
pd.read_csv('firstcsv.csv', sep = ';')
foodstuff = pd.read_csv('firstcsv.csv', sep = ';')
```

```python
plt.plot(foodstuff.Date, foodstuff.Quantity)
plt.plot(foodstuff.Date, foodstuff.Cash_Price)
plt.plot(foodstuff.Date, foodstuff.BT_Price)

plt.show()
```

```python
#Create variables
date = foodstuff.Date
quantity = foodstuff.Quantity
cash_price = foodstuff.Cash_Price
bt_price = foodstuff.BT_Price

#plt.plot(date, quantity, '*', label = 'quantity(kg)')
plt.figure(figsize = (9,7), dpi = 300)

plt.plot(date, cash_price, 'x', label = 'cash price')
plt.plot(date, bt_price, 'o', label = 'bank transfer')

plt.title('Price comparison of Potato')
plt.xlabel('date')
plt.ylabel('Price(uzs)')
plt.legend()

plt.savefig('Potatoprice.jpg')
plt.show()
```
